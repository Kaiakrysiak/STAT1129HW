```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array(["Andy", "Amy", "James", "Jules", "Arthur"])
y = np.array([88, 66, 90, 55, 77])

plt.title("Grades")
plt.bar(x,y)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np 

y = np.array([88, 66, 90, 55, 77])

plt.pie(y)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np 

y = np.array([88, 66, 90, 55, 77])
mylabels = ["Andy", "Amy", "James", "Jules", "Arthur"]

plt.pie(y, labels = mylabels)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np 

y = np.array([88, 66, 90, 55, 77])
mylabels = ["Andy", "Amy", "James", "Jules", "Arthur"]

plt.pie(y, labels = mylabels)
plt.legend(title="Grades", bbox_to_anchor=(1.05, 1.0), loc='upper left')
plt.show()
```
